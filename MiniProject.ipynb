```python
import re
import matplotlib.pyplot as plt

def detect_adversarial_text(text_array):
    """
    Detect suspicious patterns in an array of text strings that may indicate adversarial content.

    Args:
        text_array (list of str): The input array of strings to be analyzed.

    Returns:
        list of float: A list of suspiciousness scores, each between 0 and 1, with 1 being highly suspicious.
    """
    suspicious_scores = []
    total_patterns = 5  # Update this if more patterns are added

    for text in text_array:
        suspicious_score = 0

        # Pattern 1: Frequent unusual punctuation or symbols
        unusual_punctuation = re.findall(r'[!@#$%^&*()_+={}\[\]:;"\'<>?,./|]', text)
        if len(unusual_punctuation) > 3:
            suspicious_score += 0.2

        # Pattern 2: Obfuscated words (like spaces or extra characters within words)
        obfuscated_words = re.findall(r'(\w)\s+(\w)', text)  # checks for spaces within words
        if obfuscated_words:
            suspicious_score += 0.2

        # Pattern 3: Repeated characters (e.g., "soooo good" or "baaaad")
        repeated_characters = re.findall(r'(\w)\1{2,}', text)
        if repeated_characters:
            suspicious_score += 0.2

        # Pattern 4: Phishing-like phrases (you can expand this list as needed)
        phishing_phrases = ["click here", "urgent action required", "password reset"]
        if any(phrase in text.lower() for phrase in phishing_phrases):
            suspicious_score += 0.2

        # Pattern 5: All uppercase words (indicating a possible attempt to highlight or emphasize)
        uppercase_words = re.findall(r'\b[A-Z]{2,}\b', text)
        if len(uppercase_words) > 3:
            suspicious_score += 0.2

        # Calculate final score as a ratio
        suspiciousness_ratio = suspicious_score / total_patterns
        suspicious_scores.append(round(suspiciousness_ratio, 2))

    return suspicious_scores


# Example 2: Business Communication and Casual Texts
sample_texts = [
    "Congratulations! You've been selected for a FREE vacation. Claim it NOW!",  # High suspiciousness
    "Your account has been suspended due to suspicious activity. Please act immediately!",  # High suspiciousness
    "Hey, can we meet tomorrow? I have some updates to share with you.",  # Low suspiciousness
    "Your friend tagged you in a post! Check it out before it's too late.",  # Moderate suspiciousness
    "Special Offer: Get 25% off on all products this weekend only! Don't miss out!",  # Moderate suspiciousness
    "Reminder: Please submit your quarterly report by the end of the day."  # Low suspiciousness
]

suspiciousness_scores = detect_adversarial_text(sample_texts)

# Output the suspiciousness scores
for i, score in enumerate(suspiciousness_scores):
    print(f"Text {i+1} Suspiciousness Score: {score}")
    if score > 0.5:
        print(f"Warning: Text {i+1} appears suspicious and might be adversarial.")
    else:
        print(f"Text {i+1} seems safe.")

# Plotting the suspiciousness scores
plt.figure(figsize=(10, 6))
plt.bar(range(1, len(suspiciousness_scores) + 1), suspiciousness_scores, color='green', alpha=0.7)
plt.title('Suspiciousness Scores of Texts')
plt.xlabel('Text Sample')
plt.ylabel('Suspiciousness Score')
plt.xticks(range(1, len(suspiciousness_scores) + 1), [f"Text {i+1}" for i in range(len(suspiciousness_scores))], rotation=45)
plt.ylim(0, 1)
plt.grid(True)
plt.show()
```
